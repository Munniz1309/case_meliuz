```python
# Calculando intervalos dos dados para eliminar outliers
Qmin = data[['compradores',	'comissão',	'cashback',	'vendas totais']].quantile(0.15)
Qmax = data[['compradores',	'comissão',	'cashback',	'vendas totais']].quantile(0.85)

# Calculando o IQR (Intervalo Interquartil)
IQR = Qmax - Qmin
limite_inferior = Qmin - 1.5 * IQR
limite_superior = Qmax + 1.5 * IQR

# Removendo os outliers
data_sem_outliers = (data[['compradores',	'comissão',	'cashback',	'vendas totais']] >= limite_inferior) & (data[['compradores',	'comissão',	'cashback',	'vendas totais']] <= limite_superior)
data = data[data_sem_outliers.all(axis=1)].reset_index()
```
